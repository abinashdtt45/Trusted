```python
import pandas as pd
import numpy as np 
import random 

data_frame = pd.read_csv('amazon_reviews.txt', delimiter="\t")
train_data_init = data_frame['REVIEW_TEXT'].values.tolist()
train_labels_init = data_frame['LABEL'].values.tolist()

#shuffling the dataset
shuffle = list(zip(train_data_init, train_labels_init))
random.shuffle(shuffle)
train_data_init, train_labels_init =zip(*shuffle)
del data_frame

final_train_data = np.asarray(train_data_init).tolist()
final_train_labels = np.asarray(train_labels_init).tolist()

#print(final_train_labels)

for i in range(0,len(final_train_labels)):
    if(final_train_labels[i]=="__label1__"):
        final_train_labels[i]=0
    else:
        final_train_labels[i]=1

#print(final_train_labels)
print('Total training reviews ', len(final_train_data))
print('Total training labels ', len(final_train_labels))
```

```python
import nltk
sets =set()
maxx = 0
for reviews in final_train_data:
    words = nltk.tokenize.word_tokenize(reviews)
    for word in words:
        sets.add(word)
    if maxx<len(words):
        maxx= len(words)
        
print(maxx)
```

```python
len(sets)
nltk.download('punkt')
nltk.download('stopwords')
from nltk.stem import WordNetLemmatizer
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.layers import Embedding
from keras.utils import to_categorical
from nltk.corpus import stopwords
```

```python
MAX_DICT_WORDS = 55000
MAX_SEQ_LENGTH = 3500
EMBEDDING_DIM = 300
tokenizer = Tokenizer(num_words=MAX_DICT_WORDS, filters ='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~')

def tokenize_data(texts, labels):
    tokenizer.fit_on_texts(texts)
    stop = stopwords.words('english')
    tokens = [token for token in texts if token not in stop]
    word_index = tokenizer.word_index
    print('The no of unique tokens is '+ str(len(word_index)))
    lmtzr = WordNetLemmatizer()
    tokens = [lmtzr.lemmatize(word) for word in tokens]
    sequences = tokenizer.texts_to_sequences(tokens)
    data = pad_sequences(sequences, maxlen = MAX_SEQ_LENGTH)
    labels = to_categorical(labels, num_classes=len(set(labels)))
    

    return data, labels, word_index

```

```python
print(final_train_labels)
```

```python
X, Y, word_index = tokenize_data(final_train_data, final_train_labels)
print(Y)
```

```python
train_data = X[:int(len(X)*0.9)]
train_labels = Y[:int(len(X)*0.9)]
test_data = X[int(len(X)*0.9):int(len(X)*0.95)]
test_labels = Y[int(len(X)*0.9):int(len(X)*0.95)]
valid_data = X[int(len(X)*0.95):]
valid_labels = Y[int(len(X)*0.95):]
```

```python
def load_embeddings(word_index, embeddingsfile='glove.6B/glove.6B.%id.txt' %EMBEDDING_DIM):
    embedding_index = {}
    f = open(embeddingsfile, 'r', encoding='utf-8')
    for line in f:
        values = line.split(' ')
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embedding_index[word] = coefs

    f.close()


    print('Found %s word vectors ' % len(embedding_index))


    embedding_matrix = np.zeros((len(word_index)+1, EMBEDDING_DIM))
    for word, i in word_index.items():
        embedding_vector = embedding_index.get(word)
        if embedding_vector is not None:
            # words not found will all be zeros
            embedding_matrix[i]=embedding_vector


    
    embedding_layer = Embedding(len(word_index)+1,
                        EMBEDDING_DIM,
                        weights=[embedding_matrix],
                        input_length =MAX_SEQ_LENGTH,
                        trainable=False)
    
    return embedding_layer
```

```python
embedding_layer = load_embeddings(word_index)
```

```python
from keras import Sequential, Model, Input
from keras.layers import Conv1D, MaxPooling1D, AveragePooling1D, Flatten, Dense, GlobalAveragePooling1D, Dropout,LSTM, LSTM, RNN, SimpleRNN, Conv2D, GlobalMaxPooling1D
from keras import callbacks
```

```python
def cnn_model(sequence_input, embedded_sequences, classes=2):
    x = Conv1D(64, 5, activation='relu')(embedded_sequences)
    x = MaxPooling1D(5)(x)
    x = Conv1D(128, 3, activation='relu')(x)
    x = MaxPooling1D(5)(x)
    x = Conv1D(256, 2, activation='relu')(x)
    x = GlobalAveragePooling1D()(x)
    x = Dense(2048, activation='relu')(x)
    x = Dropout(0.5)(x)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.5)(x)
    preds = Dense(classes, activation='softmax')(x)

    model = Model(sequence_input, preds)
    return model
```

```python
sequence_input = Input(shape=(MAX_SEQ_LENGTH,), dtype='int32')
embedded_sequences = embedding_layer(sequence_input)

model = cnn_model(sequence_input, embedded_sequences, classes=2)
model.summary()
```

```python
from keras.callbacks import ModelCheckpoint, History
filepath="model-{epoch:02d}-{val_acc:.2f}.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['acc'])
history = model.fit(train_data, train_labels,
          validation_data=(valid_data, valid_labels),
          epochs=15, batch_size=128,callbacks=callbacks_list)
```
